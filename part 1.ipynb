```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#scatter plot >- axis level
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex',style='time',size='size')
```

```python
# relplot -> figure level -> square shape
sns.relplot(tips, x='total_bill', y='tip',kind='scatter', hue='sex',style='time',size='size')
```

```python
#line plot
gap = px.data.gapminder()
df1 = gap[gap['country']=='India']
df1
```

```python
sns.lineplot(df1,x = 'year',y='lifeExp',)
```

```python
sns.relplot(df1,x = 'year',y='lifeExp', kind = 'line')
```

```python
df1 = gap[gap['country'].isin(['India','Pakistan','Brazil','Germany','China'])]
df1
```

```python
sns.relplot(kind='line',data=df1, x='year', y='lifeExp',hue='country',style='continent',size='continent')
```

```python
# facet plot-- for multiple plot 
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter',row = 'smoker', col = 'sex')
```

```python
# col wrap

sns.relplot(data = gap , x ='year',y= 'gdpPercap',col='year',col_wrap=4)
```

```python
# distribution plot


sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x = 'total_bill',kind='hist')
```

```python
sns.displot(data=tips,x = 'total_bill',kind='hist',bins=10)
sns.displot(data=tips,x = 'tip',kind='hist',hue='sex')
```

```python
t1= sns.load_dataset('titanic')
t1
```

```python
sns.displot(data=t1,x = 'age',kind='hist')
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True)
```

```python
# Rugplot

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# heatmap
h1 = gap.pivot(index = 'country',columns = 'year',values = 'lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(h1)
```

```python
h1 = gap[gap['continent']== 'Europe'].pivot(index = 'country',columns = 'year',values = 'lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(h1,annot = True,linewidth=0.6)
```
